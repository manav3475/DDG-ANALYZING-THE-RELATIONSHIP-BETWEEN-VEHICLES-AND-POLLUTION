```python
import pandas as pd
import numpy as np

#Pre-processing
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#Plots and graphs
%cd E:\M.Sc(DS & SA)\Sem-2\Data Driven Governance\Final Project Data\Final Datasets
```

```python
df = pd.read_csv("Data_PMSO2NO2_2011-2020.csv");
df
```

```python
df_so2 = pd.DataFrame()
df_so2["Year"] = df["Year"]
df_so2["State"] = df["State"]
df_so2["SO2"] = df["SO2"]
df_so2
```

```python
le = LabelEncoder()
state = le.fit_transform(df_so2["State"])
df_so2["state_encoded"] = state
df_so2
```

```python
df_so2.drop('State',axis=1,inplace=True)
```

```python
df_2015 = df_so2.loc[df_so2["Year"]==2015]
df_2016 = df_so2.loc[df_so2["Year"]==2016]

test_2015_2016 = df_2015.append(df_2016)
test_2015_2016
```

```python
train_2011_14 = df_so2.head(132)
train_2011_14
```

```python
#x_train,x_test,y_train,y_test = train_test_split(train_2011_14,test_2015_2016,test_size=0.3,random_state=0)
x_train = train_2011_14.drop("SO2",axis=1)
y_train = train_2011_14["SO2"].copy()
x_test = test_2015_2016.drop("SO2",axis=1)
y_test = test_2015_2016["SO2"].copy()
```

```python
lr = LinearRegression()
lr_fir = lr.fit(x_train,y_train)
pred = lr.predict(x_test)
pred
```

```python
test_2015_2016["SO2"] = pred
test_2015_2016
```

```python
test_2015_2016.to_csv("pred_2015_2016_so2.csv", index = False)
```
