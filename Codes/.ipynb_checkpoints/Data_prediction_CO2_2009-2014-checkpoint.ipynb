```python
df = pd.read_csv("Data_CO2-1995-2014.csv");
df
```

```python
df_co2 = pd.DataFrame()
df_co2["Year"] = df["Year"]
df_co2["State"] = df["State"]
df_co2["CO2"] = df["CO2"]
df_co2
```
