```python
import pandas as pd
import numpy as np

#data split and model train test
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#set path of current dict
%cd E:\M.Sc(DS & SA)\Sem-2\Data Driven Governance\Final Project Data\Final Datasets
```

```python
df = pd.read_csv("Data_PMSO2NO2_2011-2020.csv");
df
```

```python
df_so2 = pd.DataFrame()
df_so2["Year"] = df["Year"]
df_so2["State"] = df["State"]
df_so2["SO2"] = df["SO2"]
df_so2
```

```python
le = LabelEncoder()
state = le.fit_transform(df_so2["State"])
df_so2["state_encoded"] = state
df_so2
```

```python
df_so2.drop('State',axis=1,inplace=True)
```

```python
df_2019 = df_so2.loc[df_so2["Year"]==2019]
df_2020 = df_so2.loc[df_so2["Year"]==2020]

test_2019_2020 = df_2019.append(df_2020)
test_2019_2020
```

```python
train_2011_18 = df_so2.head(264)
train_2011_18
```

```python
x_train = train_2011_18.drop("SO2",axis=1)
y_train = train_2011_18["SO2"].copy()
x_test = test_2019_2020.drop("SO2",axis=1)
y_test = test_2019_2020["SO2"].copy()
```

```python
lr = LinearRegression()
lr_fir = lr.fit(x_train,y_train)
pred = lr.predict(x_test)
pred
```

```python
test_2019_2020["SO2"] = pred
test_2019_2020
```

```python
test_2019_2020.to_csv("pred_2019_2020_so2.csv", index = False)
```
